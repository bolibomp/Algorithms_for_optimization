**Exercise 7.2.**  Design an objective function and a starting point $x_0$ such that Hooke-Jeeves will fail to attain a reduction in the objective function. You need to choose $x_0$ such that it is not a local minimizer.

**Answer:** $f(x) = 1-x^2/2+x^4/24$ and $x_0 = 0$ also assuming that the step size is $2\sqrt{3}$.

```python
def fourthorder(x):
    return 1-x**2/2+x**4/24

x = np.linspace(-4,4,1000)
roots = [-2*np.sqrt(3),0, 2*np.sqrt(3)]

plt.plot(x,fourthorder(x))
plt.hlines(1,roots[0],roots[-1], colors='k')
plt.scatter(roots,fourthorder(np.array(roots)), color='red')
plt.show()
```
